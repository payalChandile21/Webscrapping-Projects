```python
import requests
from bs4 import BeautifulSoup
import csv
import time

try:
    url = 'https://www.amazon.in/Apple-iPhone-11-Pro-256GB/product-reviews/B07XVMJF2D/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
    page = requests.get(url)
except Exception as e:
    error_type, error_obj, error_info = sys.exc_info()
    print('Error for link: ', url)
    print(error_type, 'line:', error_info.tb_lineno)

time.sleep(2)
soup = BeautifulSoup(page.text, 'html.parser')
links = soup.find_all('div', attrs={'class': 'a-size-base'})

data = []
for link in links:
    data.append(link.text.strip())

# Writing data to CSV
csv_file = 'amazon_reviews.csv'
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Review"])
    writer.writerows([[item] for item in data])

print(f"Scraped data has been saved to '{csv_file}'")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the Amazon page you want to scrape
url = "https://www.amazon.in/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find and extract data for all products
product_data = []
product_links = soup.find_all('a', {'class': 'a-link-normal'})
for link in product_links:
    product_url = link.get('href')
    if product_url and "/dp/" in product_url:
        product_url = "https://www.amazon.in" + product_url

        # Extract product title
        product_title_element = link.find('span', {'class': 'a-size-medium'})
        product_title = product_title_element.text.strip() if product_title_element else "N/A"

        # Extract product price
        product_price_element = link.find('span', {'class': 'a-offscreen'})
        product_price = product_price_element.text.strip() if product_price_element else "N/A"

        product_data.append([product_title, product_price, product_url])

# Export data to CSV
csv_filename = "amazon_products.csv"
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["Product Title", "Product Price", "Product Link"])
    csv_writer.writerows(product_data)

print("Data exported to", csv_filename)
```
