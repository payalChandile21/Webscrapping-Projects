```python
import requests
from bs4 import BeautifulSoup

# URL of the Amazon page you want to scrape
url = "https://www.amazon.in/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find and print the title of the page
title = soup.title.text
print("Title:", title)

# Find and print the text of a specific element (e.g., the first product title)
product_title = soup.find('span', {'class': 'a-size-medium'})
if product_title:
    print("Product Title:", product_title.text)
else:
    print("Product title not found.")

# Find and print the prices of a specific element (e.g., the first product price)
product_price = soup.find('span', {'class': 'a-offscreen'})
if product_price:
    print("Product Price:", product_price.text)
else:
    print("Product price not found.")
```
